## Enkele tips voor de data-preparatie

- Je moet drie datasets aan elkaar koppelen tot één geheel. Let er op dat je dat op de juiste manier doet: 
  - Koppel op de juiste velden.
  - Kijk na hoeveel lijnen je overhoudt na koppeling en interpreteer dit aantal. Kan het kloppen? 
  - Lees goed de beschikbare documentatie i.v.m. de weergegevens (weather.csv) en de zonnepanelen (solar.csv).
- Gebruik de laatste versies van solar.csv en weather.csv die je vindt op Chamilo. Hoe meer data, hoe beter voor je model. 
- Gebruik je gezond verstand om te bepalen welke attributen belangrijk kunnen zijn voor de opbrengst van zonnepanelen. Aarzel niet om zelf berekende attributen toe te voegen. 
- Maak enkel gebruik van de technieken voor data-preparatie en modellen die gezien zijn in de cursus. Ga dus niet op zoek naar geavanceerde technieken voor tijdsreeksen. Die heb je niet nodig om een goed voorspellingsmodel te kunnen opstellen. 
- Data voorbereiden en exploreren, modellen fine-tunen en interpreteren doe je in een Python-notebook, zodat je tijdens de presentatie gemakkelijk kan tonen welke stappen je doorlopen hebt en welke outputs je kreeg (er zal geen tijd zijn om te hertrainen). 
- Zoals elke IOT-toepassing kan de input data fouten bevatten, zoals ontbrekende gegevens en outliers. Spoor die op en zoek een verantwoorde manier om ermee om te gaan.
